```python
import pandas as pd
```

```python
df=pd.read_excel('Customer Call List.xlsx')
```

```python
df= df.drop_duplicates()
df
```

```python
df = df.drop(columns='Not_Useful_Column')
df
```

```python
df['Last_Name']=df['Last_Name'].str.strip('123/._')
df
```

```python
df["Phone_Number"]=df["Phone_Number"].astype(str).str.replace("[^a-zA-Z0-9]","",regex=True)
df
```

```python
df["Phone_Number"]=df["Phone_Number"].apply(lambda x:f'{x[0:3]}-{x[3:6]}-{x[6:10]}')df
```

```python
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('na--','')
df
```

```python
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--','')
df
```

```python
#df[['Street_Address','State','Zip_code']]=df['Address'].str.split(',',n=2,expand=True)
#df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(",", n=2, expand = True)
SplitAddress= df["Address"].str.split(" ", n=2, expand = True)
SplitAddress
```

```python
df["Street_Address"] = SplitAddress[0]
df["State"] = SplitAddress[1]
df["Zip_Code"] = SplitAddress[2]
```

```python
df
```

```python
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
```

```python
df['Paying Customer']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Paying Customer']=df['Do_Not_Contact'].str.replace('No','N')
```

```python
df
```

```python
df.fillna('')
```

```python
df.dropna(subset=['Phone_Number'], inplace=True)
df
```
