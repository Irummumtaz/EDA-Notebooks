Encoding

```python
df_object=df.select_dtypes(include='object')
df_object
```

```python
#One-Hot Incoding Label-Encoding
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
ordinal_columns=['NObeyesdad']
nominal_columns=['Gender','CALC','FAVC','SCC','SMOKE','family_history_with_overweight','CAEC','MTRANS']

#one-hot encoding
onehot_encoder = OneHotEncoder(sparse=False)
# Apply one-hot encoding to nominal columns
nominal_encoded = onehot_encoder.fit_transform(df[nominal_columns])
df_nominal_encoded = pd.DataFrame(nominal_encoded, columns=onehot_encoder.get_feature_names_out(nominal_columns))

```

```python
#Label Encoding
label_encoder = LabelEncoder()

df_ordinal_encoded = df[ordinal_columns].apply(label_encoder.fit_transform)
```

```python
df_encoded = pd.concat([df.drop(columns=nominal_columns + ordinal_columns), df_nominal_encoded, df_ordinal_encoded], axis=1)
```

```python
df=df_encoded
df
```

Standard scalar

```python
from sklearn.preprocessing import StandardScaler
continuous_cols = ['Age', 'Height', 'Weight']
scaler = StandardScaler()
scaler.fit(df[continuous_cols])
# Transform your continuous data using the fitted scaler
df_scaled_continuous = scaler.transform(df[continuous_cols])
df_scaled_continuous = pd.DataFrame(df_scaled_continuous, columns=continuous_cols)
```

```python
# Drop columns from df_encoded and concatenate with df_scaled_continuous
df_concat = pd.concat([df_encoded.drop(['Age', 'Height', 'Weight'], axis=1), df_scaled_continuous], axis=1)

df_concat
```

```python
df=df_concat
df
```

Model Development

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
X=df.drop(columns=['NObeyesdad'])
y=df[['NObeyesdad']].values
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=40)
```

```python
# Reshape target variables to 1D arrays
y_train = np.ravel(y_train)
y_test = np.ravel(y_test)
```

```python
decisiontree_model = DecisionTreeClassifier(random_state=42)
svm_model=SVC(random_state=42)
```

```python
decisiontree_model.fit(X_train,y_train)
svm_model.fit(X_train,y_train)
```

```python
y_pred_dt=decisiontree_model.predict(X_test)
y_pred_svm=svm_model.predict(X_test)
```

```python
accuracy_dt_model=accuracy_score(y_test,y_pred_dt)
accuracy_svm_model=accuracy_score(y_test,y_pred_svm)
```

```python
print(f" Decision Tree Model Accuracy: {accuracy_dt_model*100}%")
print(f" SVM Model Accuracy: {accuracy_svm_model*100}%")
```

Confusion Matrix and Classification Report

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_dt = confusion_matrix(y_test, y_pred_dt)
confusion_svm = confusion_matrix(y_test, y_pred_svm)
```

```python
# Print confusion matrices
print("\nConfusion Matrix - Decision Tree:")
print(confusion_dt)
print("\nConfusion Matrix - SVM:")
print(confusion_svm)
```

```python
# Print classification reports
print("\nClassification Report - Decision Tree:")
print(classification_report(y_test, y_pred_dt))
print("\nClassification Report - SVM:")
print(classification_report(y_test, y_pred_svm))
```
