```python
# remove some columns as they are useless 
df = df.drop(['takeaway','opentable_support',
                 'delivery', 'currency'], axis = 1)
```

Missing Values Handling

```python
missing_values_total = df.isnull().sum().sum()
print(f"Total missing values: {missing_values_total}")
# Analyze missing values in each column
missing_values_per_column = df.isnull().sum()
print("\nMissing values in each column:")
print(missing_values_per_column)
```

```python
missing_data_perc = df.isnull().sum() / len(df) * 100 #Percentage of missing values
missing_data_perc 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out columns with 0% missing values to declutter the visualization
missing_data_filtered = missing_data_perc[missing_data_perc > 0]

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=missing_data_filtered.index, y=missing_data_filtered, palette='coolwarm')
plt.xticks(rotation=90) 
plt.ylabel('Percentage of Missing Values')
plt.title('Missing Data Analysis per Column')
plt.show()
```

Given that the 'zipcode' column has 79.3% missing data and attempting to impute these values would compromise the dataset's integrity, it's advisable to drop the 'zipcode' column.

```python
df.drop(['zipcode'] , inplace = True , axis = 1 )
```

```python
print(f"Shape after dropping some columns {df.shape}")
```

```python
df['address'] = df['address'].fillna('Unknown') # fill missing address with unknown
df['address'].isnull().sum()
```

```python
# Calculate the mode value (most frequent values) of the 'cuisines' column
cuisine_mode_value = df['cuisines'].mode()[0]  # Get the first mode value
df['cuisines'] = df['cuisines'].fillna(cuisine_mode_value)
df['cuisines'].isnull().sum()
```

```python
def fill_missing_timings(group): #replace the missing values with restaurant name 
    try:
        mode_value = group.mode().iloc[0]
        return group.fillna(mode_value)
    except IndexError:
        return group

df['timings'] = df.groupby('name')['timings'].transform(fill_missing_timings)
```

```python
df['timings'].isnull().sum() #remaining missing value 
```

```python
cuisine_mode_value = df['timings'].mode()[0]  # Get the first mode value
df['timings'] = df['timings'].fillna(cuisine_mode_value)
df['timings'].isnull().sum()
```
